This Notebook contains the analysis and required graphing for Regnet models

```python
import re
import matplotlib.pyplot as plt
```

```python
def parse_log(log_file):
    epochs = []
    top1_accuracy = []
    top5_accuracy = []

    with open(log_file, 'r') as f:
        for line in f:
            if 'Epoch(val)' in line:
                epoch_match = re.search(r'\[([0-9]+)\]', line)
                top1_match = re.search(r'accuracy_top-1: ([0-9.]+)', line)
                top5_match = re.search(r'accuracy_top-5: ([0-9.]+)', line)
                if epoch_match and top1_match and top5_match:
                    epochs.append(int(epoch_match.group(1)))
                    top1_accuracy.append(float(top1_match.group(1)))
                    top5_accuracy.append(float(top5_match.group(1)))

    return epochs, top1_accuracy, top5_accuracy

def plot_performance(epochs, accuracy, model_name, metric):
    plt.plot(epochs, accuracy, label=f'{model_name} - {metric}')
```

Add Files to Plot in the list below

```python
# List of log files
log_files = ['data_regnetx1.6.log',
             'data_regnetx3.2.log',
             'data_regnetx4.0.log',
             'data_regnetx6.4.log',
             'data_regnetx8.0.log',
             'data_regnetx12.log',
             'data_regnetx400mf.log',
             'data_regnetx800mf.log'
]
             
```

```python
# Plotting Top-5 Accuracy
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
for log_file in log_files:
    epochs, _, top5_accuracy = parse_log(log_file)
    model_name = log_file.split('_')[1].split('.')[0].upper()  # Extract model name from file name
    plot_performance(epochs, top5_accuracy, model_name, 'Top-5 Accuracy')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Top-5 Accuracy Comparison')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting Top-1 Accuracy
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
for log_file in log_files:
    epochs, top1_accuracy, _ = parse_log(log_file)
    model_name = log_file.split('_')[1].split('.')[0].upper()  # Extract model name from file name
    plot_performance(epochs, top1_accuracy, model_name, 'Top-1 Accuracy')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Top-1 Accuracy Comparison')
plt.legend()
plt.grid(True)
plt.show()
```
